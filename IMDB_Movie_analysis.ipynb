###Movies Based on Reviews good Average excelent

```python
def review(rating):
  if rating>=7.0:
    return 'Good'
  elif rating>=6.0:
    return 'Average'
  else:
    return 'Excelent'
```

```python
df['Reviwes']=df['Rating'].apply(review)
df.head(10)
```

```python
df[df['Reviwes']=='Excelent'].head(3)
```

### unique Genre

```python
df['Genre'].unique()
```

```python
list1=[]
for value in df['Genre']:
  list1.append(value.split(','))
list1[0:6]
```

```python
listU=[]
for item in list1:
  for item1 in item:

      listU.append(item1)
listU
```

```python
ulist=[]
for it in listU:
  if it not in ulist:
    ulist.append(it)
```

```python
print('Unique Genere',ulist)
```
